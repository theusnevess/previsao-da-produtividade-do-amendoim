# Modelagem - Regressão Linear Múltipla

```python
X = df.loc[:, "temp_SEP":] # Features -> Variáveis independentes
y = df.loc[:, "produtividade"] # Target -> Variável dependente 

X
```

```python
# Padronização e normalização dos dados 
from sklearn.preprocessing import StandardScaler

escala = StandardScaler().fit(X) # Instanciação da classe de padronização, ajustando pelos valores de X
Xpad = escala.transform(X) # Aplica o valor da coluna subtraído pela sua média e dividido pelo seu desvio padrão
Xpad = pd.DataFrame(Xpad, columns=[df.columns[4:]])
Xpad
```

```python
# Importando as configurações do modelo de regressão linear
from sklearn.linear_model import LinearRegression
from mlxtend.feature_selection import SequentialFeatureSelector as sfs

# Instanciando o modelo 
modelo = LinearRegression()

# Selecionando as variáveis utilizadas pelo modelo, reduzindo assim a dimensionalidade para k features
sfs_select = sfs(modelo, k_features=20, forward=True, scoring="r2", cv= 3)

# Treinando o modelo 
sfs_select = sfs_select.fit(Xpad, y)
sfs_select
```

```python
# Convertendo os dados estatísticos da seleção de características para um dataframe 
tabela = pd.DataFrame.from_dict(sfs_select.get_metric_dict()).T
tabela
```

```python
from mlxtend.plotting import plot_sequential_feature_selection as plt_sfs 

# Plotando o gráfico da seleção de características 
fig1 = plt_sfs(sfs_select.get_metric_dict(), kind="std_dev")
plt.grid()
```

```python
# Exibindo quais são as 10 melhores features para o modelo 
tabela.loc[10, "feature_names"]
```

```python
import statsmodels.api as sm

# Criando um dataframe com as 10 melhores features 
Xobs = Xpad[[
    'ue_ANN',
    'tmax_SEP',
    'tmax_DEC',
    'tamplitude_NOV',
    'Prec_SEP',
    'Prec_DEC',
    'Qg_SEP',
    'Qg_DEC',
    'Qg_ANN',
    'ETP_OCT'
]]
yobs = y

# Adicionando coluna constante (ainda não ajustada)
Xobs = sm.add_constant(Xobs)
```

```python
# Ajuste do modelo 
resultado = sm.OLS(yobs, Xobs).fit()

# Imprimindo o valor dos pesos (coeficientes)
resultado.summary()
```
